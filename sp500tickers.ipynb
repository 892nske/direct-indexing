```python
import pandas as pd

# 現在のティッカー一覧が格納されたDataFrameを作成
current_tickers = pd.DataFrame({'ticker': ['AAPL', 'GOOG', 'AMZN', 'FB', 'MSFT', 'JPM', 'BAC', 'WFC', 'C', 'V']})

# ティッカー一覧の推移を出力する関数を定義
def get_ticker_history(current_tickers, ticker_changes):
    # ティッカー追加/削除情報をDataFrameに変換
    ticker_changes = ticker_changes[['date', 'ticker_add', 'ticker_remove']]
    ticker_changes = ticker_changes.melt(id_vars='date', value_name='ticker')
    ticker_changes = ticker_changes.drop('variable', axis=1)
    
    # ティッカー一覧を日付ごとに格納するDataFrameを作成
    ticker_history = pd.DataFrame({'date': pd.date_range(ticker_changes['date'].min(), ticker_changes['date'].max())})
    ticker_history['tickers'] = None
    
    # 日付ごとにティッカー一覧を更新
    for date in ticker_history['date']:
        # 当該日付までに追加/削除されたティッカーを抽出
        changes_until_date = ticker_changes[ticker_changes['date'] <= date]
        tickers_added = set(changes_until_date['ticker_add'].dropna())
        tickers_removed = set(changes_until_date['ticker_remove'].dropna())
        
        # 追加/削除されたティッカーを反映して、当該日付のティッカー一覧を作成
        current_tickers_set = set(current_tickers['ticker'])
        tickers_set = current_tickers_set.union(tickers_added) - tickers_removed
        tickers_list = sorted(list(tickers_set))
        tickers_str = ', '.join(tickers_list)
        ticker_history.loc[ticker_history['date'] == date, 'tickers'] = tickers_str
    
    return ticker_history

# テスト用の入力データを作成
current_tickers = pd.DataFrame({'ticker': ['AAPL', 'GOOG', 'AMZN', 'FB', 'MSFT', 'JPM', 'BAC', 'WFC', 'C', 'V']})
ticker_changes = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03'],
    'ticker_add': ['TSLA', 'NVDA', None],
    'security_add': ['Tesla Inc', 'Nvidia Corp', None],
    'ticker_remove': [None, None, 'WFC'],
    'security_remove': [None, None, 'Wells Fargo & Co']
})

# ティッカー一覧の推移を出力
ticker_history = get_ticker_history(current_tickers, ticker_changes)
print(ticker_history)
```
